```python
import numpy as np

# Datos brutos (una sola lista larga que nos llegó mal del sistema)
# El orden es repetitivo: Altura, Puntos, Edad, Altura, Puntos, Edad...
raw_data = [190, 15, 25, 185, 22, 21, 210, 8, 30, 175, 30, 24, 
            201, 12, 28, 198, 18, 26, 182, 25, 22, 215, 5, 31, 
            192, 14, 23, 188, 20, 25]

# Convertimos a array de numpy
players_array = np.array(raw_data)
```

Primero ordenamos los datos mediante el uso de reshape para formar las columnas que ya emos definido anteriormente ( altura, puntos y edad )

```python
team_matrix=players_array.reshape(10,3)
print(team_matrix)
```

```python
altura,puntos,edad=np.hsplit(team_matrix,3)

ages=np.copy(edad)
print(ages)
#Mi solucion esta bien pero puede causar un poco de molestias si trato de pasar los datos a un modelo de ML pues estos por lo general prefieren datos planos y no de mas dimensiones.
#podria haber hecho el slicing sin necesidad de usar split: ages= team_matrix[:,2].copy
```
