**Comparado y Filtrado Condicional** 

Tienes el ID de  clientes que calificaron para un credito de dos meses 

**RETO** Usa las ufuncs de operaciones de conjuntos para encontrar:

Los clientes que calificaron en ambos meses.

Los clientes que son nuevos en el mes 2 (que no estaban en el mes 1).

```python
import numpy as np 
mes1=np.array([101,102,103,104,105])
mes2=np.array([103,104,105,106,107])

cl_cali=np.intersect1d(mes1,mes2)
print("Los clientes aprovados en ambos meses son: ", cl_cali)
cl_nw2=np.setdiff1d(mes2,mes1)

print("Los nuevos clientes en el segundo mes son: ", cl_nw2)

```
