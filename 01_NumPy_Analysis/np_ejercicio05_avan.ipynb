# **Análisis de Ingresos "Pareto"**
**Reto:**
Genera dos grupos de ingresos (arreglos) de 50 personas cada uno usando una distribución de Pareto con un parámetro $a=3$.

**Operación Basic:**

Une ambos grupos en un solo arreglo de 100 personas (Join/Concatenate).

Cambia el tipo de dato (Data Type) a int para que los ingresos no tengan decimales.

Busca la posición (Search) del ingreso más alto en el arreglo (puedes usar np.where o argmax si lo conoces).


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
```

```python
grupo01=random.pareto(a=3,size=50)
grupo02=random.pareto(a=3,size=50)
```

```python
total_group=np.concatenate((grupo01,grupo02),axis=0)
total_group=total_group.astype(int)
print(total_group)
```

```python
maximo=np.max(total_group)
datosmax=np.where(total_group==maximo)
print(datosmax)
```
