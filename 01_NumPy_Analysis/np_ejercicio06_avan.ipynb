# **Encuesta de Satisfacción**

**Reto** : Simula los resultados de 50 personas donde el 20% está disconforme, el 30% neutro y el 50% conforme, usando una Distribución Multinomial.

**Operación Basic**:

El resultado será un arreglo con los conteos. Usa un bucle (Iterating) con nditer para imprimir un mensaje por cada categoría, algo como: "Categoría X: Total Y votos".

Divide el experimento en 2 grupos de 25 personas usando array_split.

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random 
```

```python
encuesta= random.choice([0,1,2],size=50,p=[0.2,0.3,0.5])
print(encuesta)
```

```python
categorias = ["Disconforme", "Neutro", "Conforme"]

conteos = np.bincount(encuesta, minlength=3)

i = 0
for total in np.nditer(conteos):
    print(f"Categoría {i} ({categorias[i]}): Total {total} votos")
    i += 1
```

```python
group1,group2=np.array_split(encuesta,2)
print(group1,group2)
```
