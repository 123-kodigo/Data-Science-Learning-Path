# **Control de Calidad en la Fábrica**
**Reto**: Genera 100 datos de pesos usando una Distribución Normal ($\mu=150, \sigma=2$).

**Operación Basic**:

Ordena los pesos de menor a mayor (Sort).

Imprime solo los 5 componentes más livianos y los 5 más pesados usando indexación negativa.

Crea una copia (Copy) del arreglo y reemplaza todos los valores que estén fuera del rango $[146, 154]$ por el número 0 (estos son los productos fallados).


```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from numpy import random
```

```python
fabrica=random.normal(loc=150,scale=2,size=100)
print(fabrica)
```

```python
fabrica=np.sort(fabrica)
print(fabrica)
```

```python
m_weight=fabrica[:5]
l_weight=fabrica[-5:]
print(m_weight)
print(l_weight)
```

```python
fabrica_cop=np.where((fabrica >=146)&(fabrica<=154),fabrica,0)
print(fabrica_cop)
```
