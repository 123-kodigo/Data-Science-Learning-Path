**RETO**: Genera un arreglo que represente las llamadas recibidas por minuto durante una hora (60 minutos) usando una distribución de Poisson con un promedio
 ($\lambda$) de 4 llamadas por minuto.

**Operación Basic**:

Corta el arreglo (Slicing) para ver solo los primeros 15 minutos.

Cambia la forma (Reshape) del arreglo original a una matriz de (6, 10) (donde cada fila es una decena de minutos).

Usa un filtro (Filter) para encontrar en qué minutos se recibieron más de 7 llamadas (posible colapso del sistema).

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from numpy import random
```

```python
calls=random.poisson(lam=4,size=60)
print(calls)
```

```python
#Mostrar unicamente los 15 primeros numeros
fift=calls[:15] 

#Modificamos la forma del array ahora es de 6 filas y 10 columnas
calls_rs=calls.reshape(6,10)
print(calls_rs)
#Obtenemos los datos cuando sean mayor a 7 usando where
fila,columna=np.where(calls_rs >7)
for i,j in zip(fila,columna):
    print(calls_rs[i,j])
```
