**NORMALIZACION DE DATOS**

Tienes las edades de un grupo de estudiantes

**Reto**: Utiliza ufuncs para aplicar la fórmula de normalización Min-Max: 


```python
import numpy as np 

edades=np.array([18,21,25,30,22,24])

minimos_valor=np.min(edades)
maximo_valor=np.max(edades)
#Se usa de fondo el substract y divide para restar los valores como minimos_valor sin necesidad de aplicar un loop que recorra cada elemento del array
#Esto  ya viene optimisado por C 
xnormal=(edades-minimos_valor)/(maximo_valor-minimos_valor)
print(xnormal)
```
