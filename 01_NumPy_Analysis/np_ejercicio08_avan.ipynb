**Análisis de Crecimiento Logarítmico**

Tienes un saldo de una cuenta de ahorros que crece de forma exponencial.

**RETO** Usa la ufunc de logaritmo natural para transformar los datos y luego usa np.diff() para encontrar la tasa de crecimiento entre cada periodo.

```python
import numpy as np 
ahorros=np.array([100,150,225,337.5,506.25])

normal=np.log(ahorros)

tasa=np.diff(normal)
print(tasa)
```
